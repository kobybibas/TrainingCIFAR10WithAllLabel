```python
import numpy as np
import matplotlib.pyplot as plt
import json
from analyze_utilities import *
from dataset_utilities import create_cifar10_dataloaders
```

```python
# Open results files

# New Alg.
json_file_name_NML = '../output/NML_results_20180810_140920/results_NML_20180810_140920.json'
json_file_name_CIFAR100 = '../output/OutOfDist_CIFR100_results_20180810_153726/results_OutOfDist_CIFR100_20180810_153726.json'
json_file_name_Noise = '../output/OutOfDist_Noise_results_20180811_100246/results_OutOfDist_Noise_20180811_100246.json'

with open(json_file_name_NML) as data_file:    
    results_dict_NML = json.load(data_file)
with open(json_file_name_CIFAR100) as data_file:    
    results_dict_CIFAR100 = json.load(data_file)
with open(json_file_name_Noise) as data_file:    
    results_dict_Noise = json.load(data_file)
```

```python
# Extract and plot normalization factor
normalization_factor_CIFAR10_list = get_normalization_factor_from_dict(results_dict_NML)
normalization_factor_CIFAR100_list = get_normalization_factor_from_dict(results_dict_CIFAR100)
normalization_factor_Noise_list = get_normalization_factor_from_dict(results_dict_Noise)
print('log10(NormalizationFactor): [cifar10 cifar100 Noise]=[%f %f %f] ' % 
      (np.mean(normalization_factor_CIFAR10_list), 
       np.mean(normalization_factor_CIFAR100_list), 
       np.mean(normalization_factor_Noise_list)))
plt.figure(figsize=(20,10))
bins= 50
plt.hist(np.log10(normalization_factor_CIFAR10_list), alpha=1, label='CIFAR 10 Image', density=True)
plt.hist(np.log10(normalization_factor_CIFAR100_list), alpha=0.5,  label='CIFAR 100 Image', density=True)
plt.hist(np.log10(normalization_factor_Noise_list), alpha=0.5,  label='Noise Image', density=True)
plt.ylabel('Count', fontsize=18)
plt.xlabel('log(NormalizationFactor)', fontsize=18)
plt.title('Histogram of log(normalization factor)', fontsize=22)
plt.grid()
plt.xlim([0,1])
plt.legend()
plt.show()
```

```python
loss_NML_list, acc_NML = get_NML_log_loss_from_dict(results_dict_NML)
loss_ERM_list, acc_ERM = get_ERM_log_loss_from_dict(results_dict_NML)
loss_Jinni_list, acc_Jinni = get_jinni_log_loss_from_dict(results_dict_NML)

print('[NML ERM Jinni]: Acc=[%f %f %f], Loss=[%f %f %f]' % 
      (acc_NML, acc_ERM, acc_Jinni, np.mean(loss_NML_list), np.mean(loss_ERM_list), np.mean(loss_Jinni_list)))


plt.figure(figsize=(20,10))
bins= 20
plt.hist(loss_NML_list, alpha=1, label='NML', density=True)
plt.hist(loss_ERM_list, alpha=0.5,  label='ERM', density=True)
plt.hist(loss_Jinni_list, alpha=0.5,  label='Jinni', density=True)
plt.ylabel('Count', fontsize=18)
plt.xlabel('Logloss', fontsize=18)
plt.title('Logloss Histogram', fontsize=22)
plt.grid()
plt.xlim([0,1])
plt.legend()
plt.show()
```
